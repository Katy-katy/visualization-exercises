## Object Oriented Matplotlib

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)


# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

### subplots()

The plt.subplots() object will act as a more automatic axis manager.

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# We can specify the number of rows and columns when creating the subplots() object:
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# axex is an array
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object    
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'g')
axes[1].plot(y, x, 'r')
    
axes[0].set_title('title 1')
axes[1].set_title('title 2')

fig    
plt.tight_layout() # to adjust the positions of the axes on the figure canvas so that there is no overlapping content
# or fig.tight_layout()
```

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.plot(y, x, 'g')#the second graph on the same figure
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12,3))

axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'g')

axes[0].set_title('title 1');
axes[1].set_title('title 2');

plt.tight_layout()
```

```python
fig = plt.figure() 
ax = fig.add_axes ([0,0,1,1])

ax.plot(x, x**2, label = "X **2")
ax.plot(x, x**3, label = "X **3")

ax.legend(loc=0) # best location (10 for center) or we can: loc = (0.1, 0.1)
```

```python
fig = plt.figure() 
ax = fig.add_axes ([0,0,1,1])

ax.plot(x, y, color = "#F88C00", lw = 10, alpha = 0.4, ls = "--", marker = 'o', 
        markersize = 20, markerfacecolor = "red", markeredgewidth=3, markeredgecolor="green") 
# color RGB HEX code
# possible linestype options -, , -., :, steps
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
```

```python
fig = plt.figure() 
ax = fig.add_axes ([0,0,1,1])

ax.plot(x, y, color = "g", lw = 5, ls = ":") 
ax.set_xlim([0, 1])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles**

```python
fig = plt.figure() 
ax = fig.add_axes ([0,0,1,1])

ax.plot(x, y) 

ax.set_xlabel('X Label') # Notice the use of set_ to begin methods
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

** Create the plot by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, z, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(x, y, 'r')
axes2.set_xlim([20, 22])
axes2.set_ylim([30, 50])
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Zoom');
```

** Use plt.subplots(nrows=1, ncols=2) to create a plot. Now plot (x,y) and (x,z) on the axes.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, color = "#F88C00", lw = 5, alpha = 0.4, ls = "--")
axes[1].plot(x, z, color = "#F88C11", lw = 2, alpha = 0.9, ls = ":")
    
axes[0].set_title('title 1')
axes[1].set_title('title 2')

fig    
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,3))

axes[0].plot(x, y, color = "#F88C00", lw = 5, alpha = 0.4, ls = "--")
axes[1].plot(x, z, color = "#F88C11", lw = 2, alpha = 0.9, ls = ":")
    
axes[0].set_title('title 1')
axes[1].set_title('title 2')

fig    
plt.tight_layout()
```
